**IRIS Flower Classification**

```python
import numpy as np
import matplotlib as mpl
import pandas as pd
import seaborn as sb
from sklearn.model_selection import train_test_split
column=['sepal length','sepal width','petal length','petal width','class labels']
df=pd.read_csv('/content/IRIS.csv',names=column, header=0)
df.head(150)
```

This cell imports essential libraries for data analysis and visualization (NumPy, Matplotlib, Pandas, Seaborn) and the train_test_split function from scikit-learn. It defines column names for the Iris dataset and loads the CSV file with these column names. Finally, it displays the first 150 rows of the dataset (which is the entire dataset).

```python
df.describe()
```

This cell generates descriptive statistics of the dataset's numerical columns, showing count, mean, standard deviation, minimum, 25th percentile, median (50th percentile), 75th percentile, and maximum values for each feature (sepal length, sepal width, petal length, petal width).

```python
#ploting data to know the relation
sb.pairplot(df,hue='class labels')
```

This cell creates a pairplot using Seaborn to visualize relationships between all pairs of features. The hue='class labels' parameter colors the data points according to their species classification. This visualization helps identify patterns and clustering among the three Iris species across different feature combinations.

```python
data=df.values
Y=data[:,4]
X=data[:,0:4]
print(X)
```

 This cell converts the DataFrame to a NumPy array and splits it into features (X) and target (Y). X contains the four features (sepal length, sepal width, petal length, petal width), while Y contains the class labels. It then prints the features matrix X to show all the measurement values.

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
print((X_train),len(X_test), (Y_train),len(Y_test))
```

This cell splits the dataset into training and testing sets with 80% of data used for training and 20% for testing. It prints the training feature set, the length of the test set (30 samples), the training target values, and confirms the test set length (30 samples).

```python
from sklearn.svm import SVC
obj=SVC()
obj.fit(X_train,Y_train)
prediction=obj.predict(X_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,prediction))
```

This cell implements a Support Vector Classification model. It imports the SVC class, instantiates an SVC object, trains it on the training data, makes predictions on the test data, and then calculates and prints the accuracy score (0.9666 or 96.66%).

```python
# using regression model
from sklearn.linear_model import LogisticRegression
obj1= LogisticRegression()
obj1.fit(X_train,Y_train)
prediction1=obj1.predict(X_test)
```

This cell implements a Logistic Regression model as an alternative classification approach. It imports the LogisticRegression class, creates an instance, trains it on the training data, and generates predictions for the test data.

```python
print(accuracy_score(Y_test,prediction1))
```

This cell calculates and prints the accuracy score for the Logistic Regression model (0.9666 or 96.66%), which matches the accuracy of the SVM model in this case.
